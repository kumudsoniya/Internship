```python
#add a column for ranking by happiness score
df['DystopiaResidual']= df.index + 1
```

```python
# create a random column called 'DystopiaResidual'
df['DystopiaResidual'] = 1.5 + 1.5*np.random.randn(len(df))
```

```python
cor = df.corr()
cor
```

```python
# Plot the correlation matrix
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#split the dataset into features(x) and target (y)
X = df.drop(['Country','Happiness Score'], axis=1)
y = df['Happiness Score']
```

```python
df
```

```python
#split the data into training and testing sets
X_train,X_test , y_train, y_test =train_test_split(X, y, test_size =0.2, random_state=7)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg= RandomForestRegressor(n_estimators=100,random_state =42)
rf_reg.fit(X_train,y_train)
```

```python
#prediction on test set 
from sklearn.metrics import mean_squared_error
y_pred=rf_reg.predict(X_test)
rmse= np.sqrt(mean_squared_error(y_test,y_pred))
print(f'RMSE: {rmse}')
```

```python
 #scatter plot to visualize the model's predictions.
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('True Values')
plt.ylabel('prediction')
plt.show()
```

```python
#save the trained model for future
import joblib
joblib.dump(rf_reg, 'World happiness_model.pkl')
```
